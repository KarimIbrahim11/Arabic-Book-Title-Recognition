```python
from PIL import Image, ImageDraw
from IPython.display import display

# Get bounding boxes
output = results.xywhn[0].cpu().numpy()
print(output)
bboxes = output[:, :4]
print(bboxes)
confidences = output[:, 4]
classes = output[:, 5]

# rounded_tensor = torch.round(bbox).cpu()
# print(rounded_tensor)
```

```python
def bbox_retrieval(bbox, W_img, H_img):
  x, y , w ,  h = bbox
  x1 = (x - w/2) * W_img
  y1 = (y - h/2) * H_img
  x2 = (x + w/2) * W_img
  y2 = (y + h/2) * H_img
  return (x1, y1, x2, y2)

def crop_text_regions(bbox, image):
  x1, y1, x2, y2 = bbox
  cropped_image = image.crop(bbox)
  return cropped_image
```

```python
import numpy as np
# Open the image and create ImageDraw object for drawing
image = Image.open(inf_image)
width, height = image.size
draw = ImageDraw.Draw(image)

cropped_images = []
bbox_points = []
# Draw the bounding box with red lines
for idx, bbox in enumerate(bboxes):
  # Filter only labels with class 0 aka Title
  if classes[idx] == 0:
    x1,y1,x2,y2 = bbox_retrieval(bbox,width,height)
    bbox_points.append([x1,y1,x2,y2])

    draw.rectangle((x1,y1,x2,y2), outline=(255, 0, 0), width=5)

# Display image with Title bounding boxes
display(image)
```

```python
sorted_bounding_boxes = sorted(bbox_points, key=lambda bbox: bbox[1])
for bbox in sorted_bounding_boxes:
  cropped_image = crop_text_regions(tuple(bbox), image)
  cropped_images.append(cropped_image)
  display(cropped_image)
```
